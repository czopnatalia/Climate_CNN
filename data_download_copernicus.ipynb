## Tworzenie requestów do pobrania danych

```python
import cdsapi

dataset = "reanalysis-era5-land"

c = cdsapi.Client(url="https://cds.climate.copernicus.eu/api",
                        key="f85a23bc-7b23-4cea-92fa-ec62fa5810fc")

krakow_area = [50.2, 19.7, 49.9, 20.2]

for year in range(2000, 2026): 
    print(f"Rozpoczynam pobieranie danych dla roku {year}...")
    
    filename = f"krakow_temp_{year}.nc"
    
    try:
        c.retrieve(
            dataset,
            {
                "variable": ["2m_temperature"],
                "year": str(year),
                "month": [f"{m:02d}" for m in range(1, 13)],
                "day": [f"{d:02d}" for d in range(1, 32)],
                "time": [f"{h:02d}:00" for h in range(24)], 
                "data_format": "netcdf",
                "area": krakow_area,
            }
        ).download(filename)
        print(f"Zakończono: {filename}")
    
    except Exception as e:
        print(f"Błąd przy pobieraniu roku {year}: {e}")
```
